## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
def fit_line_robust(xs, ys, ws):
    ''' use iteratively reweighted least squares to fit the equation y = mx+b
    input is the x and y pixel coordinates and the weight (image intensity) for each pixel.
    
    This is basically a transcription of the Wikipedia article into numpy.
    '''
    p = 1  # l_p norm
    # Initial reweighted weights are all 1
    W = np.ones(xs.shape)
    X = np.vstack([xs, np.ones(xs.shape)]).T.astype(np.float32)
    ys = ys.astype(np.float32)
    del xs  # we are now done with xs, free it
    # X.T is [[x1 x2 x3 ... xn]
    #            1  1  1      1]]
    b0 = None
    itr = 0
    while itr < 20:
        Wt = np.diag(W * ws)
        XTW = np.dot(X.T, Wt)
        beta = np.linalg.solve(np.dot(XTW, X) + np.diag([1e-5, 1e-5]), np.dot(XTW, ys))
        if b0 is not None:
            e = (beta - b0) / b0
        if b0 is not None and np.dot(e, e) < 1e-5:
            return beta
        b0 = beta
        W = np.maximum(1e-6, np.abs(ys - np.dot(X, beta))) ** (p-2)
        itr += 1

    return beta

def estimate_lane_lines(edgeimage):
    '''
    For each pixel in the region of interest, consider it part of a least squares line
    fit problem. We'll use iteratively reweighted least squares with an l_p minimization
    for more robust estimation than least squares would give us ordinarily.
    
    For each pixel in the edge image which is not black, if it's on the left half of the
    screen, consider it part of the left lane; otherwise consider it part of the right lane.
    
    We'll run a separate optimization for each, and then return the tuple
    ([left_m, left_b], [right_m, right_b]) for the slope and intercept of left and right lines.
    '''
    center = edgeimage.shape[1] / 2
    nzy, nzx = np.nonzero(edgeimage)  # x and y coordinates of all Canny edge pixels
    left_x = nzx[nzx < center - 20]  # x, y coordinates of all edge pixels on the left
    left_y = nzy[nzx < center - 20]
    left_w = edgeimage[(left_y, left_x)]
    right_x = nzx[nzx > center + 20]  # ...and right
    right_y = nzy[nzx > center + 20]
    right_w = edgeimage[(right_y, right_x)]
    # Note: I am intentionally flipping x and y here, as I actually want to fit x = my + b
    # as it's more convenient to work with y as the independent variable
    leftmb = fit_line_robust(left_y, left_x, left_w)
    rightmb = fit_line_robust(right_y, right_x, right_w)
        
    return leftmb, rightmb
    
    
def estimate_lane_lines_hough(image, gray):
    ''' Compute an estimate for the slope/intercepts of the left lane and right lane
    in screen space. Returns an array of [sum(m), sum(b), N] for each of left and right
    lanes; sum(m)/N and sum(b)/N are the mean slope and intercept estimate, respectively.
    
    NOTE: This is not used in my final solution.
    '''

    
    rho = 2
    theta = np.pi / 90
    threshold = 50
    min_line_length = 5
    max_line_gap = 10

    lines = cv2.HoughLinesP(gray, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)

    ''' Let's make some simplifying assumptions... we want to find the left lane marker
    and the right lane marker, and all lines we detect are either left, right, or neither.
    Left lane markers have dy/dx < -thresh, and right lane markers have dy/dx > +thresh
    
    Let's average the slope and offset parameters we see for each of those; we'll weight
    the average by line length so small outliers don't affect us as much.
    
    Also, we'll be departing from convention slightly by using x = my + b, as we want to use
    y as our independent coordinate and find x as y varies from the bottom of the screen to
    the middle.
    '''
    
    laneparams = np.zeros((2, 3))  # sum(m), sum(b), N
    if lines is not None:
        for line in lines:
            for x1, y1, x2, y2 in line:
                if y1 == y2:
                    continue
                dy = y2 - y1
                dx = x2 - x1
                length = np.sqrt(dx**2 + dy**2)
                m = float(dx) / dy
                b = x1 - float(dx*y1) / dy
                
                if m < -0.5:
                    laneparams[0] += np.array([m*length, b*length, length])
                elif m > 0.5:
                    laneparams[1] += np.array([m*length, b*length, length])
                #if np.abs(y1 - y2) > 10:
                #    continue
                cv2.line(image, (x1, y1), (x2, y2), (255, 0, 0, 128), 1)
    return laneparams
    

def annotate_lanes(image, vp_y=310, bottom_y=540):
    ''' Detect the lane markers in the image, and draw virtual green lane markers on them.
    
    vp_y is the y coordinate of the virtual vanishing point in the image, and bottom_y
    is the bottom row of the image to consider (mostly to reject car hoods).'''
    h, w, _ = image.shape
    # Grayscale, blur, canny edge detect
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    gray = cv2.GaussianBlur(gray, (5, 5), 0)
    gray = cv2.Canny(gray, 100, 200)

    # For our region of interest, assume the trapezoid +/- 30 pixels from the
    # center at the vanishing point, down to the bottom_y.
    roi_vertices = np.array([[(100, bottom_y), (w/2 - 30, vp_y), (w/2 + 30, vp_y), (w - 100, bottom_y)]], np.int32)
    gray = region_of_interest(gray, roi_vertices)

    # plt.imshow(gray, 'gray')
    # plt.show()

    leftmb, rightmb = estimate_lane_lines(gray)
    # use x = m*y + b to compute locations of the left and right lanes
    x1 = int(leftmb[0] * (h-1) + leftmb[1])
    x2 = int(leftmb[0] * vp_y + leftmb[1])
    cv2.line(image, (x1, h - 1), (x2, vp_y), (0, 255, 0, 128), 5)
    x1 = int(rightmb[0] * (h-1) + rightmb[1])
    x2 = int(rightmb[0] * vp_y + rightmb[1])
    cv2.line(image, (x1, h - 1), (x2, vp_y), (0, 255, 0, 128), 5)
    
    return image

IMAGEDIR = "test_images"
for file in os.listdir(IMAGEDIR):
    fname = os.path.join(IMAGEDIR, file)
    plt.imshow(annotate_lanes(mpimg.imread(fname)))
    plt.show()
````
